```python
# Plot: Precipitation daily average with rolling average
fig, ax = plt.subplots(figsize=(16,4))
ax = plots.setup_month_axes(ax)

window = 15 # days

ax.bar(ddf.index, ddf['average'], width=1)
ax.plot(ddf.index, ddf['average'].rolling(window, center=True, win_type='triang').mean(), alpha=0.7, color='r')

plt.ylabel('Evaporation (mm)')
plt.legend(['Moving average ({} days)'.format(window),'Daily average (2000-2017)'])

plt.savefig(site.path_outputdir+'/EvapMA.jpg')
```

Calculate daily stats (mean and std) for selected pool and variable

```python
df = tools.daily_stats(simpool, var)
```

```python
plt.figure(figsize=(20,8))
plt.bar(df.index, df.average, width=1)
plots.setup_month_axes(plt.gca())
```

```python
plt.figure(figsize=(20,8))
plt.bar(df.index, df.stdev, width=1)
plots.setup_month_axes(plt.gca())
```

```python
mean = df.average
std = df.stdev
factor = 0.5
lowci = [x-y*factor if x-y*factor>0 else 0 for x,y in zip(mean,std)]
highci = [x+y*factor if x+y*factor <1 else 1 for x,y in zip(mean,std)]
ax = plots.lineplotCI(df.index, mean, df.index, lowci, highci, '',name_var, '')
plots.setup_month_axes(ax)
ax.legend(['Daily average (2000-2017)', 'Avg. $\pm$ 0.5*std'])
plt.savefig(site.path_outputdir+'/depthpercent_pattern.jpg')
```
